```python
from ultralytics import YOLO
import cv2

# Load the YOLOv8 model
model = YOLO('yolov8m.pt')

# Initialize counts for cars moving in and out of the city
cars_in_city = 0
cars_out_city = 0

# Load the video file
video_path = '/content/car1.mp4'
cap = cv2.VideoCapture(video_path)

# Define video writer to save the output video
output_path = "/content/final_output.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, 30.0, (int(cap.get(3)), int(cap.get(4))))

# Define a reference line in the middle of the frame (for cars moving in/out)
frame_height = int(cap.get(4))
middle_line = frame_height // 2

while cap.isOpened():
    ret, frame = cap.read()
    # The if statement should be aligned with the while loop
    if not ret:
        break

    # Run YOLOv8 inference
    results = model(frame)

    # Draw bounding boxes and track cars
    for result in results:
        boxes = result.boxes.cpu().numpy()  # Get the bounding boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Extract box coordinates
            confidence = box.conf[0]  # Get the confidence score
            cls = int(box.cls[0])  # Get the class ID

            # Only consider cars (YOLO's class for car is typically 2)
            if cls == 2 and confidence > 0.5:
                # Calculate the center of the bounding box
                center_y = (y1 + y2) // 2

                # Check if the car is moving in or out of the city
                if center_y < middle_line:
                    color = (0, 255, 0)  # Green for cars moving in
                    cars_in_city += 1
                else:
                    color = (0, 0, 255)  # Red for cars moving out
                    cars_out_city += 1

                # Draw the bounding box with the respective color
                cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

                # Display the counts on the frame
                cv2.putText(frame, f'Cars In: {cars_in_city}', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                cv2.putText(frame, f'Cars Out: {cars_out_city}', (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    # Write the modified frame to the output video
    out.write(frame)

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()
```
